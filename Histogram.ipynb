### Write a Python program using OpenCV to perform histogram equalization on the given grayscale image **"parrot.jpg"**.

The program should:

- Import the required libraries (**OpenCV, NumPy, Matplotlib**).
- Read the image **"parrot.jpg"** in **grayscale** format.
- Display the grayscale image and plot its histogram.
- Perform **histogram equalization** on the grayscale image using `cv2.equalizeHist()` to enhance its contrast.
- Using Matplotlib, display the following in a **2 × 2 grid layout**:
  - Original Grayscale Image
  - Histogram of Original Image
  - Enhanced (Equalized) Image
  - Histogram of Enhanced Image


```python
# Import required libraries
import cv2
import matplotlib.pyplot as plt

```

```python
# Read the image in grayscale format
img_grayscale=cv2.imread("parrot.jpg", cv2.IMREAD_GRAYSCALE)

```

```python
# Display the grayscale image.
plt.imshow(img_grayscale,cmap="gray")

```

```python
# Plot the histogram of the grayscale image

grayHist = cv2.calcHist(img_grayscale, [0], None, [256], [0,256])
# plt.figure(figsize=(10,10))
plt.hist(img_grayscale.ravel(), 256, range=[0,256])

plt.show()
```

```python
# Perform histogram equalization
hist_eq = cv2.equalizeHist(img_grayscale)
plt.imshow(hist_eq, cmap="gray")
```

```python
# Display [1] the Original Image (Gray Image) and its Histogram, and [2] the Enhanced Image and its Histogram using a 2×2 layout in Matplotlib.
plt.figure(figsize=(12,8))



plt.subplot(221)
plt.imshow(img_grayscale,cmap="gray")
plt.title("Original Image")


plt.subplot(222)
plt.hist(img_grayscale.ravel(), 256, range=[0,256])
plt.title("Histogram of Original Image")

plt.subplot(223)
plt.imshow(hist_eq, cmap="gray")
plt.title("Enhanced Image")

plt.subplot(224)
plt.hist(hist_eq.ravel(), 256, range=[0,256])
plt.title("Histogram of Enhanced Image")
```

### Write a Python program using OpenCV to perform color image enhancement through histogram equalization in the HSV color space on the given image **"parrot.jpg"**.

The program should:

- Import the required libraries (**OpenCV, NumPy, Matplotlib**).
- Read the color image **"parrot.jpg"**.
- Plot the **histogram of the original color image** (use the **B, G, and R** channels).
- Convert the image from **BGR to HSV** color space.
- Apply **histogram equalization** to one specific channel to improve brightness and contrast using `cv2.equalizeHist()`.
- Convert the enhanced HSV image back to **BGR** format.
- Using Matplotlib, display the following in a **2 × 2 grid layout**:
  - Original Color Image
  - Histogram of Original Image (BGR channels)
  - Enhanced Color Image
  - Histogram of Enhanced Image (BGR channels)


```python
# Read the colorgiven parrot.jpg image.

parrot_bgr = cv2.imread("parrot.jpg")
parrot_rgb = cv2.cvtColor(parrot_bgr, cv2.COLOR_BGRA2RGBA)

```

```python
# Plot the histogram of colour image

plt.hist(parrot_rgb.ravel(), 256, range=[0,256])
plt.show()

```

```python
# Convert to HSV.

parrot_hsv = cv2.cvtColor(parrot_rgb, cv2.COLOR_RGB2HSV)


```

```python
# Perform histogram equalization

H, S, V = cv2.split(parrot_hsv)
hist_eq = cv2.equalizeHist(V)
hist_eq_hsv = cv2.merge((H, S, hist_eq))
```

```python
# Convert back to BGR format

hist_eq_bgr = cv2.cvtColor(hist_eq_hsv, cv2.COLOR_HSV2BGR)


```

```python
# Display

plt.imshow(cv2.cvtColor(hist_eq_bgr, cv2.COLOR_BGR2RGB))


```

```python
# Display [1] the Original Image (BGR Image) and its Histogram, and [2] the Enhanced Image and its Histogram using a 2×2 layout in Matplotlib.
plt.figure(figsize=(12,8))



plt.subplot(221)
plt.imshow(cv2.cvtColor(parrot_bgr, cv2.COLOR_BGR2RGB))
plt.title("Original Image")


plt.subplot(222)
plt.hist(parrot_bgr.ravel(), 256, range=[0,256])
plt.title("Histogram of Original Image")

plt.subplot(223)
plt.imshow(cv2.cvtColor(hist_eq_bgr, cv2.COLOR_BGR2RGB))
plt.title("Enhanced Image")

plt.subplot(224)
plt.hist(hist_eq_bgr.ravel(), 256, range=[0,256])
plt.title("Histogram of Enhanced Image")
```
